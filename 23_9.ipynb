```python
import folium
import pandas as pd
from sklearn.cluster import KMeans
from geopy.distance import geodesic
goc = [10.7565577, 106.6248006]
khach = [[10.7532397, 106.6335908], [10.7551474, 106.6342982], [10.7500, 106.6200], [10.7600, 106.6400], [10.7450, 106.6250]]
m = folium.Map(location=goc, zoom_start=14)
folium.Marker(location=goc, popup="Goc", icon=folium.Icon(color='red', icon='home')).add_to(m)
for i in khach: folium.Marker(location=i, popup='Khach hang', tooltip='Khach Hang').add_to(m)
model = KMeans(n_clusters=2, random_state=42).fit(khach)
ds_kho = model.cluster_centers_
for kho in ds_kho: folium.Marker(location=kho, popup="KHO HANG TOI UU", icon=folium.Icon(color='green', icon='star')).add_to(m)
folium.Circle(location=goc, radius=1500, color='green', fill=True, fill_opacity=0.2).add_to(m)
m
```
