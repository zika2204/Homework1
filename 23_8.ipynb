```python
import folium
from geopy.distance import geodesic
nha=[10.7371309, 106.6755775]
khach=[{"ten":"A","c":[10.7371309, 106.6755775]},{"ten":"B","c":[10.7608154, 106.6632871]}]
xe=[{"id":"1","c":[10.7830141, 106.6948927]},{"id":"2","c":[10.952106, 106.8638217]},{"id":"3","c":[10.7500,106.6200]}]
m=folium.Map(location=nha,zoom_start=13)
for i in khach:
    folium.Marker(i["c"],popup=i["ten"],icon=folium.Icon(color='red')).add_to(m)
    x=min(xe,key=lambda v:geodesic(i["c"],v["c"]).meters)
    folium.Marker(x["c"],popup=x["id"],icon=folium.Icon(color='green')).add_to(m)
    folium.PolyLine([i["c"],x["c"]]).add_to(m)
m
```
