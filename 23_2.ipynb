```python
import folium
from geopy.distance import geodesic
from geopy.geocoders import Nominatim
geolocator = Nominatim(user_agent="my_exercise_app")
place = [
    "Dinh Doc Lap, Ho Chi Minh City",
    "UEH Campus A, Nguyen Dinh Chieu, Ho Chi Minh City",
    "Parc Mall, Ta Quang Buu, Ho Chi Minh City",
    "Pham Ngoc Thach Hospital, Ho Chi Minh City",
    "Ben Thanh Market, Ho Chi Minh City",
    "Thao Cam Vien Zoo, Ho Chi Minh City",
    "Thong Nhat Stadium, Ho Chi Minh City",
    "Van Hanh Mall, Ho Chi Minh City",
    "Dam Sen Park, Ho Chi Minh City",
    "Marie Curie High School, Ho Chi Minh City"
]
toa_do = []

for i in place:
    location = geolocator.geocode(i)
    if location:
        toa_do.append((i, (location.latitude, location.longitude)))
        print(f'Ten: {location.address}')
        print(f'Toa do: {location.latitude}, {location.longitude}')
    else:
        print(f"Không tìm thấy tọa độ cho: {i}")
for a in toa_do[0:]:
    point_a = (10.7371309, 106.6755775)
    ten_dia_diem = a[0]
    point_b = a[1]
    distance = geodesic(point_a, point_b).km
    print(f'Khoang cach thuc te: {distance:.2f} km')
m = folium.Map(location=point_a, zoom_start=14)
point_a = (10.7371309, 106.6755775)
ten_diem_goc = "My home"
folium.Marker(
    location=point_a,
    popup=ten_diem_goc
).add_to(m)
for b in toa_do:
    ten_dia_diem = b[0]
    point_b = b[1]
    distance = geodesic(point_a, point_b).km
    folium.Marker(
        location=point_b,
        popup=f"{ten_dia_diem}: {distance:.2f} km",
        icon=folium.Icon(color='blue')
    ).add_to(m)
    folium.PolyLine(
        locations=[point_a, point_b],
        weight=2,
        color='blue',
        tooltip=f"{distance:.2f} km"
    ).add_to(m)
m
```
