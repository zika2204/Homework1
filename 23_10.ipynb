```python
import folium
import random
goc = [10.7769, 106.7009]
m = folium.Map(location=goc, zoom_start=14)
place = [
    [10.7725, 106.6980, "Ngã sáu Quách Thị Trang", 90, 5],
    [10.7827, 106.6925, "Ngã tư Nam Kỳ Khởi Nghĩa", 85, 8],
    [10.7748, 106.7019, "Đường Đồng Khởi", 40, 35],
    [10.7712, 106.6901, "Ngã sáu Phù Đổng", 95, 4],
    [10.7905, 106.7052, "Cầu Thị Nghè", 75, 12]
]
def phantich(mat_do, toc_do):
    if mat_do > 80 and toc_do < 10:
        return "Rất cao", "red"
    elif mat_do > 60 or toc_do < 20:
        return "Trung bình", "orange"
    else:
        return "Thấp", "green"
for loc in place:
    lat, lon, name, density, speed = loc
    level, color1 = phantich(density, speed)
    folium.Circle(
        location=[lat, lon],
        radius=200,
        color=color1,
        fill=True,
        fill_opacity=0.4,
        popup=f"{name}<br>Nguy cơ: {level}<br>Mật độ: {density}% - Tốc độ: {speed}km/h"
    ).add_to(m)

duong_thay_the = [
    [10.7830, 106.6948],
    [10.7860, 106.6980],
    [10.7790, 106.7050]
]
folium.PolyLine(
    locations=duong_thay_the,
    color="blue",
    weight=5,
    opacity=0.8,
    dash_array='10',
    tooltip="Tuyến đường thay thế đề xuất (Né vùng tắc nghẽn)"
).add_to(m)
folium.Marker(duong_thay_the[0], popup="Điểm đi").add_to(m)
folium.Marker(duong_thay_the[-1], popup="Điểm đến", icon=folium.Icon(color='green')).add_to(m)
m
```
