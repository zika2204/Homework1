```python

import folium
import osmnx as ox
from sklearn.cluster import MeanShift
home = [10.7830141, 106.6948927]
khach = [[10.78301, 106.69427], [10.78031, 106.694489], [10.78411, 106.69423], [10.7800, 106.6900], [10.7850, 106.6950], [10.7880, 106.6800]]
G = ox.graph_from_point(home, dist=500, network_type='drive')
nodes, edges = ox.graph_to_gdfs(G)
model = MeanShift(bandwidth=0.005).fit(khach)
station = model.cluster_centers_
m = folium.Map(location=home, zoom_start=15)
folium.GeoJson(edges, style_function=lambda x: {'color':'gray', 'weight':1, 'opacity':0.3}).add_to(m)
for a in khach:
  folium.CircleMarker(location=a, radius=3, color='blue').add_to(m)
for b in station:
  folium.Marker(location=b, icon=folium.Icon(color='green', icon='flash')).add_to(m)
m
```
