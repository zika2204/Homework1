```python

import folium
from geopy.distance import geodesic
m = folium.Map(location=[10.7830141, 106.6948927], zoom_start=16)
nhan = [[10.78301, 106.69427], [10.78031, 106.694489], [10.78411, 106.69423]]
duong = [[10.7830141, 106.6948927]] + sorted(nhan, key=lambda x: geodesic([10.7830141, 106.6948927], x).meters)
l1 = folium.FeatureGroup(name='Store').add_to(m)
folium.Marker([10.7830141, 106.6948927],icon=folium.Icon(color='pink')).add_to(m)
folium.Marker([10.78301, 106.69427],icon=folium.Icon(color='green')).add_to(m)
folium.Marker([10.78031, 106.694489],icon=folium.Icon(color='blue')).add_to(m)
folium.Marker([10.78411, 106.69423],icon=folium.Icon(color='red')).add_to(m)
l2 = folium.FeatureGroup(name='Service Area').add_to(m)
folium.Circle([10.7830141, 106.6948927], radius=600, fill=True).add_to(l2)
folium.LayerControl().add_to(m)
folium.PolyLine(duong, color='yellow', weight=5).add_to(m)
m
```
