```python

import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
from shapely.geometry import Polygon

data = {
    'District': ['Bien Hoa', 'Long Xuyen', 'Long Thanh', 'Tri Ton'],
    'geometry': [
        Polygon([(0,0), (1,0), (1,1), (0,1)]), # Vùng 1
        Polygon([(1,0), (2,0), (2,1), (1,1)]), # Vùng 2
        Polygon([(0,1), (1,1), (1,2), (0,2)]), # Vùng 3
        Polygon([(1,1), (2,1), (2,2), (1,2)])  # Vùng 4
    ]
}
gdf = gpd.GeoDataFrame(data, crs="EPSG:4326")

business_data = pd.DataFrame({
    'District': ['Bien Hoa', 'Long Xuyen', 'Long Thanh', 'Tri Ton'],
    'Revenue': [100, 50, 20, 80]
})

map_data = gdf.merge(business_data, on="District")

fig, ax = plt.subplots(1, 1, figsize=(10, 6))
map_data.plot(column='Revenue',
              cmap='OrRd',
              legend=True,
              edgecolor='black', ax=ax)
plt.title("Biểu đồ phân bổ doanh thu theo khu vực")
plt.show()
```
