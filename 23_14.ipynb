```python
import folium
from folium.plugins import TimestampedGeoJson

m = folium.Map(location=[10.77, 106.69], zoom_start=14)

xe_di_chuyen = [
    {"time": "2024-05-20T08:00:00", "coords": [10.770, 106.690]},
    {"time": "2024-05-20T08:05:00", "coords": [10.775, 106.695]},
    {"time": "2024-05-20T08:10:00", "coords": [10.780, 106.700]}
]

features = [{
    'type': 'Feature',
    'geometry': {'type': 'Point', 'coordinates': [p['coords'][1], p['coords'][0]]},
    'properties': {'time': p['time'], 'style': {'color': 'red'}}
} for p in xe_di_chuyen]

TimestampedGeoJson({'type': 'FeatureCollection', 'features': features},
                   period='PT5M', add_last_point=True).add_to(m)
m
```
