```python

import folium
from geopy.distance import geodesic
kho = [10.7830141, 106.6948927]
nhan = [[10.7371309, 106.6755775], [10.7878424, 106.7063352], [10.7801, 106.7200]]
duong = [kho] + sorted(nhan, key=lambda x: geodesic(kho, x).meters)
m = folium.Map(location=kho, zoom_start=13)
folium.Marker(kho, icon=folium.Icon(color='green', icon='home')).add_to(m)
for p in nhan: folium.Marker(p, icon=folium.Icon(color='pink')).add_to(m)
folium.PolyLine(duong, color='purple', weight=3).add_to(m)
m
```
