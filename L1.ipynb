```python
from socket import AddressFamily
import folium
from geopy.geocoders import Nominatim
import osmnx as ox
import networkx as nx
address_A = "Ben Thanh Market, Ho Chi Minh City, Vietnam"
address_B = "Tan Son Nhat Airport, Ho Chi Minh City, Vietnam"

geolocator = Nominatim(user_agent="routing-example", timeout=10)
location_A = geolocator.geocode(address_A)
location_B = geolocator.geocode(address_B)
lating_A= (location_A.latitude, location_A.longitude)
lating_B = (location_B.latitude, location_B.longitude)
midpoint = ((lating_A[0] + lating_B[0])/2,(lating_A[1] + lating_B[1])/2)
G = ox.graph_from_point(midpoint, dist=3000, network_type='drive')

orig_node = ox.distance.nearest_nodes(G, lating_A[1], lating_A[0])
dest_node = ox.distance.nearest_nodes(G, lating_B[1], lating_B[0])
# Find the shortest path by length
route = nx.shortest_path(G, orig_node, dest_node, weight='length')
# Route coordinates
route_coords = [(G.nodes[n]['y'], G.nodes[n]['x']) for n in route]
m = folium.Map(location=midpoint, zoom_start=14)
# Add markers A and B
folium.Marker(lating_A, popup="Điểm A", icon=folium.Icon(color='red')).add_to(m)
folium.Marker(lating_B, popup="Điểm B", icon=folium.Icon(color='blue')).add_to(m)
# Draw the route according to reality
folium.PolyLine(locations=route_coords, color='green', weight=5, opacity=0.8).add_to(m)
# Show map
display(m)
```
